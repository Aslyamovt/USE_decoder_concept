```python
import numpy as np

i=0
vectors = []
with open("D:/decoder/vectors_test.txt", "r", encoding='utf-8') as fr:
    for line in fr:
        a = line.replace('\n', '').split('\t')
        vectors.append(list(map(float,a)))
        i+=1
        if i==16:
            break

i=0
text=[]
with open("D:/decoder/texts_test.txt", "r", encoding='utf-8') as fr:
    for line in fr:
        a = line.replace('\n', '')
        text.append(a)
        i+=1
        if i==16:
            break

x_test = np.array(vectors)
vectors=[]
```

```python
from keras.models import Model
from keras.models import load_model
from keras.utils.generic_utils import get_custom_objects
import keras
from keras_transformer import *


from keras_layer_normalization import LayerNormalization
from keras_multi_head import MultiHeadAttention
from keras_position_wise_feed_forward import FeedForward
from keras_pos_embd import TrigPosEmbedding
from keras_embed_sim import EmbeddingRet, EmbeddingSim

model = load_model('D:/decoder/m10.h5', custom_objects = get_custom_objects ())
#model.summary()
```

```python
def binarycalculator(k):
    buf=np.zeros(shape=(10))
    out=np.zeros(shape=(512,10))
    for i in range(k):
        buf=numbertoarray(i+1)
        out[i]=out[i]+buf
    return out
    
    
def numbertoarray(m):
    out = np.zeros(shape=(10))
    powers=[512,256,128,64,32,16,8,4,2,1]
    for i in range(10):
        a=m//powers[i]
        if a>=1 and a<2:
            out[i] = 1 
            m-=powers[i]
        if m == 0:
            break
    return out 


sentences = []
k=0
for v in x_test:
    pool=[]
    a = 3001
    i=0    
    while a!=3002 and i!=512:
        piece = np.zeros(len(pool)+1)
        piece[0]=3001
        for j in range(len(pool)):
            piece[j+1]=pool[j]+1
        positions=binarycalculator(k)
        a=np.argmax(model.predict([v.reshape(1,512),positions.reshape(1,positions.shape[0],positions.shape[1]),piece.reshape(1,piece.shape[0])])[0][0])
        if a!=3002 and a!=3001 and a!=0:
            pool.append(a-1) 
        i+=1
    sentences.append(list(map(int, pool)))
    k+=1
    print(k)
```
